Acerca de:
En este dataset exploraremos los distintos tipos de datos referidos a los Datos de Comercio Electronico

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/data.csv", encoding="ISO-8859-1")
df.head()
```

Variables Categoricas:

InvoiceNo: Cada factura que genera el sistema tiene un numero unico asignado por el mismo sistema. Este numero permite rastrear cada compra y productos que pudieron haber sido vendido juntos.

StockCode: Codigo único perteneciente a cada producto del inventario.

Description: Variable que nos da una descripcion de cada producto incluido en la factura.

Country: Variable que indica el pais de residencia del cliente.

CustomerID: Identificador unico de cada cliente, el que permite hacer un rastreo de sus compras y asi proporcionar ofertas o promociones personalizadas.

Variables Numericas:

UnitPrice: Indica el valor unitario de cada producto y al ser combinada con Quantity podemos obtener el total de alguna compra.

Quantity: Variable que nos indica la cantidad de unidades de cierto producto que se incluyen en la factura.

Variables Temporales:
InvoiceDate: Variable que registra la fecha y hora en la que se realizo la transaccion.

```python
# Seleccionar las columnas numéricas relevantes
columnas_numericas = ["Quantity", "UnitPrice"]

# Calcular medidas estadísticas de tendencia central
estadisticas = df[columnas_numericas].describe().T
moda = df[columnas_numericas].mode().iloc[0]  # Obtener la moda para cada columna

# Agregar moda a la tabla de estadísticas
estadisticas["mode"] = moda

# Mostrar resultados
print(estadisticas)

```

La desviacion estandar es de 218.08 lo que nos indica que dentro de los productos comprados existe una gran variabilidad.

Analisis de Distribucion de Datos

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar el estilo de los gráficos
sns.set_style("whitegrid")

# Histograma de Quantity
plt.figure(figsize=(12, 5))
sns.histplot(df["Quantity"], bins=50, kde=True)
plt.title("Distribución de Quantity")
plt.xlabel("Cantidad de productos comprados")
plt.ylabel("Frecuencia")
plt.show()

# Histograma de UnitPrice
plt.figure(figsize=(12, 5))
sns.histplot(df["UnitPrice"], bins=50, kde=True)
plt.title("Distribución de UnitPrice")
plt.xlabel("Precio Unitario")
plt.ylabel("Frecuencia")
plt.show()

# Boxplot de Quantity
plt.figure(figsize=(12, 5))
sns.boxplot(x=df["Quantity"])
plt.title("Boxplot de Quantity")
plt.show()

# Boxplot de UnitPrice
plt.figure(figsize=(12, 5))
sns.boxplot(x=df["UnitPrice"])
plt.title("Boxplot de UnitPrice")
plt.show()
```

Distribucion de Quantity
Histograma: Podemos observar que la distribucion esta mas sesgada hacia la derecha, lo que nos indica que la mayoria de las compras fueron de pocas unidades pero aun asi existen compras de muchos productos que elevan el promedio
Boxplot: Aqui podemos observar que existen muchos valores atipicos en el extremo superior, es decir, existen transacciones que registran cantidades muy altas de productos
Distribucion de UnitPrice
Histograma: Muestra un sesgo hacia la derecha, lo que nos dice que la mayoria de los productos tienen precios bajos.
Boxplot: muchos valores en el extremo superior, es decir, algunos productos son mucho mas caros que el resto.
